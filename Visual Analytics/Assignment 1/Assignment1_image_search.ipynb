## Color histogram

```python
# Make function for plotting channels
def plot_channels (flower1):
    # Split channel
    channels = cv2.split(flower1)
    # name colors
    color_names = ["b", "g", "r"]
    # create figure
    plt.figure()
    # add title
    plt.title("Multiple channels")
    # add xlabel
    plt.xlabel("Bins")
    # add ylabel
    plt.ylabel("Number of pixels")
    for channel, color_name in zip(channels, color_names):
        # create histogram
        hist = cv2.calcHist([channel], [0], None, [256], [0,256])
        # plot histogram
        plt.plot(hist, color = color_name)
        # set limits of x axis
        plt.xlim([0, 256])

    # show the plot
    plt.show()
    
    return None

# plot histogram for image1
plot_channels(flower1)
```
